Imported required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Defined activation function

```python
def identity(x):
    return x

def bin_step(x):
    return 1 * (x >= 2) 

def bipolar_step(x):
  return 1 * (x >= 0) - 1 * (x < 0)

def bin_sigmoidal(x,l):
    return 1 / (1 + np.exp(-l*x))

def bip_sigmoidal(x,l):
    return (1 - np.exp(-l*x)) / (1 + np.exp(-l*x))
```

Plotted Identity function

```python
X = np.linspace(-5, 5, 100)
plt.plot(X, identity(X),'r')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Identity Function')
plt.grid()
plt.show()
```

Plotted Binary step function

```python
plt.plot(X, bin_step(X),'r')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Binary Step Function')
plt.grid()
plt.show()
```

Plotted Bipolar step function

```python
plt.plot(X, bipolar_step(X),'r')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Bipolar Step Function')
plt.grid()
plt.show()
```

Plotted Binary sigmoid function

```python
for l in range(1, 5):
  plt.plot(X, bin_sigmoidal(X,l),'r')
  plt.xlabel('X Axis')
  plt.ylabel('Y Axis')
  plt.title('Binary Sigmoid Function')
plt.grid()
plt.text(4, 0.8, r'$binsigmoidal(x)=\frac{1}{1+e^{-x}}$', fontsize=10)
plt.show()
```

Plotted Binary sigmoid function

```python
for l in range(1, 5):
  plt.plot(X, bip_sigmoidal(X,l),'r')
  plt.xlabel('X Axis')
  plt.ylabel('Y Axis')
  plt.title('Bipolar Sigmoid Function')
plt.grid()
plt.text(4, 0.8, r'$binsigmoidal(x)=\frac{1-e^{-x}}{1+e^{-x}}$', fontsize=10)
plt.show()
```
